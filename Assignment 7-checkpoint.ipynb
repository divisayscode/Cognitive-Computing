```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
roll_number = 102317006
np.random.seed(roll_number)
```

```python
sales_data = np.random.randint(1000, 5000, (12, 4))
```

```python
months = ["jan", "feb", "mar", "apr", "may", "jun", "jul", "aug", "sep", "oct", "nov", "dec"]
categories = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]
df = pd.DataFrame(sales_data, columns=categories, index=months)
print(df.head())
print(df.describe())
```

```python
print("first 5 rows:\n", df.head())
print("Summary :\n", df.describe())
```

```python
total_sales_category = df.sum()
total_sales_month = df.sum(axis=1)
print("Total Sales per Category:\n", total_sales_category)
print("Total Sales per Month:\n", total_sales_month)
```

```python
growth = df.diff().mean()
print("Average Growth per Category:\n", growth)
```

```python
df["Total Sales"] = df.sum(axis=1)
df["Growth Rate"] = df["Total Sales"].pct_change() * 100
print(df)
```

```python
if roll_number % 2 == 0:
    df["Electronics"] *= 0.90 
else:
    df["Clothing"] *= 0.85  
print("Discount applied:\n", df.head())
```

```python
df[categories].plot(kind='line', marker='o', figsize=(10, 5))
plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend(title="Category")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df[categories])
plt.title("Sales Distribution per Category")
plt.show()
```
