```python
import numpy as np
import matplotlib.pyplot as plt
M=float(input("Enter the value of m"))
x=np.linspace(-10,10,100)
y1=M*x*x
y2=M*np.sin(x)
plt.plot(x,y1,'b--', label='y=M*x^2')
plt.plot(x,y2,'g', label='y=M*Sin(x)')
plt.legend()
plt.grid()
plt.title("Mathematical Functions")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {'Subject': ['English', 'Math', 'Science', 'Punjabi', 'History'], 
        'Score': [90, 100, 100, 80, 96]}
df = pd.DataFrame(data)
plt.figure(figsize=(8, 5))
ax = sns.barplot(x='Subject', y='Score', data=df, hue='Subject', palette="muted", legend=False)
for bar in ax.patches:
    ax.annotate(f"{bar.get_height()}", 
                (bar.get_x() + bar.get_width() / 2, bar.get_height()), 
                ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')
plt.title("Subject-wise Scores", fontsize=14, fontweight='bold')
plt.xlabel("Subjects", fontsize=12, fontweight='bold')
plt.ylabel("Scores", fontsize=12, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
roll_no = int(input("Enter your roll number as seed: "))
np.random.seed(roll_no)
data = np.random.randn(50)
fig, ax = plt.subplots(2, 2, figsize=(10, 8))
ax[0, 0].plot(np.cumsum(data), marker='s', linestyle='-', color='g', linewidth=2)
ax[0, 0].set_title("Cumulative Sum Line Plot", fontweight='bold', fontsize=14)
ax[0, 0].set_xlabel("Index", fontsize=12)
ax[0, 0].set_ylabel("Cumulative Sum", fontsize=12)
ax[0, 0].grid(True, linestyle='--', alpha=0.6)
ax[0, 1].scatter(range(50), data, color='purple', alpha=0.7, edgecolors='black')
ax[0, 1].set_title("Scatter Plot with Random Noise", fontweight='bold', fontsize=14)
ax[0, 1].set_xlabel("Index", fontsize=12)
ax[0, 1].set_ylabel("Value", fontsize=12)
ax[0, 1].grid(True, linestyle='--', alpha=0.6)
ax[1, 0].set_visible(False)
ax[1, 1].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Divanshi\Downloads\company_sales_data.csv")
sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 5))
sns.lineplot(x=df["month_number"], y=df["total_profit"], marker='o', color='darkgreen')
plt.title("Total Profit Over All Months", fontweight='bold')
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
for column in df.columns[1:-1]:
    sns.lineplot(x=df["month_number"], y=df[column], label=column, marker='o', color='teal')
plt.title("Product Sales Over Months", fontweight='bold')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend(title="Products")
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
df.drop(columns=["month_number"]).sum().plot(kind="bar", colormap="plasma")
plt.title("Total Sales and Profit for All Features", fontweight='bold')
plt.xlabel("Attributes")
plt.ylabel("Total Sales/Profit")
plt.grid(axis='y', linestyle="--")
plt.show()
```
